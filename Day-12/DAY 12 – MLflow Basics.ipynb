```python
import mlflow
import mlflow.sklearn

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Read GOLD Delta data using PATH
gold_df = spark.read.format("delta") \
    .load("/Volumes/workspace/ecommerce/ecommerce_data/gold/products")
```

```python
# Data Preparation
# Convert to Pandas for sklearn
df = gold_df.select("views", "revenue", "purchases").dropna().toPandas()

display(df)
```

```python
# Features & target
X = df[["views", "revenue"]]
y = df["purchases"]
```

```python
# Train-test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# MLFlow Experiment Run
with mlflow.start_run(run_name="linear_regression_v1"):
    # Log parameters
    mlflow.log_param("model_type", "LinearRegression")
    mlflow.log_param("features", "views, revenue")
    mlflow.log_param("test_size", 0.2)

    # Train model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Evaluate
    score = model.score(X_test, y_test)
    mlflow.log_metric("r2_score", score)

    # Log model
    mlflow.sklearn.log_model(model, artifact_path="model")

print(f"R2 Score: {score:.4f}")
```
