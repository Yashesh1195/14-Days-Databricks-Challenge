```python
import mlflow
from transformers import pipeline

# Explicit model + revision (recommended)
MODEL_NAME = "distilbert-base-uncased-finetuned-sst-2-english"

classifier = pipeline(
    task="sentiment-analysis",
    model=MODEL_NAME
)

reviews = [
    "This product is amazing!",
    "Terrible quality, waste of money"
]

results = classifier(reviews)

print(results)

# Log to MLflow
with mlflow.start_run(run_name="sentiment_model"):
    
    mlflow.log_param("model_name", MODEL_NAME)
    mlflow.log_param("num_samples", len(reviews))

    # Example metric (confidence avg)
    avg_conf = sum(r["score"] for r in results) / len(results)
    mlflow.log_metric("avg_confidence", avg_conf)
```
