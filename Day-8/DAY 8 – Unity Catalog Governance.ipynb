```python
%sql
-- Create catalog and schemas
CREATE CATALOG IF NOT EXISTS ecommerce;
USE CATALOG ecommerce;

CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

```python
%sql
-- Register Delta tables using Volumes (FIXED)
CREATE TABLE IF NOT EXISTS bronze.events
USING DELTA
AS SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/bronze/events`;

CREATE TABLE IF NOT EXISTS silver.events
USING DELTA
AS SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/silver/events`;

CREATE TABLE IF NOT EXISTS gold.products
USING DELTA
AS SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/gold/products`;
```

```python
%sql
-- Controlled view
CREATE VIEW IF NOT EXISTS gold.top_products AS
SELECT product_id, revenue, conversion_rate
FROM gold.products
WHERE purchases > 10
ORDER BY revenue DESC LIMIT 100;

SELECT * FROM gold.top_products;
```
